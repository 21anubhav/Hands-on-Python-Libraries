**Average App Reviews**

```python
def convert_reviews(x):
    if isinstance(x, str):
        x = x.strip()
        if x.endswith('M'):
            return float(x[:-1]) * 1_000_000
        elif x.endswith('K'):
            return float(x[:-1]) * 1_000
        elif x.endswith('B'):
            return float(x[:-1]) * 1_000_000_000
        else:
            return float(x)  # plain number string
    return x  # already number

# Apply conversion
df['Reviews'] = df['Reviews'].apply(convert_reviews)

# Now calculate average
s = 0
for i in df['Reviews'].dropna():
    s += int(i)

print("Average Reviews:", int(s / len(df['Reviews'].dropna())))
```

```python
c=0
for i in df['Category']:
    if(i == "AUTO_AND_VEHICLES"):
        c+=1
print(c)
```

```python
for name in df['Category'].unique():
    ct=0
    for i in df['Category']:
        if(i==name):
            ct+=1
    print(name,':',ct)
```

 **Total number of apps in each category**

```python
categories = {}

for name in df['Category'].unique():
    ct = 0
    for i in df['Category']:
        if(i == name):
            ct += 1
    categories[name] = ct
    
print(categories)
```

**Total number of apps in each Type**

```python
types={}
for name in df['Type'].unique():
    ct=0
    for i in df['Type']:
        if(i==name):
            ct+=1
    types[name]=ct
print(types)

```

**Total number of apps in each Content Rating**

```python
content_rating = {}

for name in df['Content Rating'].unique():
    ct = 0
    for i in df['Content Rating']:
        if(i == name):
            ct += 1
    content_rating[name] = ct
    
print(content_rating)
```
