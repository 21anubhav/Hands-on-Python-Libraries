```python
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("AB_NYC_2019.csv")
```

```python
df.head(2)
```

## **Categorical**
**Bar Graph**

```python
sns.countplot( x=df["room_type"])
```

```python
sns.countplot(data=df, x= "neighbourhood_group")
```

```python
sns.countplot(data=df, x="room_type", hue="neighbourhood_group", palette="coolwarm")
```


## **Pie Chart**

```python
df["room_type"].value_counts().plot(kind = "pie", autopct = "%.2f")
```

```python
df["availability_365"].value_counts().head(10).plot(kind ="bar")
```
