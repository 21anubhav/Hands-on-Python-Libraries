```python
import pandas as pd

data = {
    "Roll_No": [101, 102, 103, 104, 105, 102, 104, 106, 107, 103],
    "Name": ["Amit", "Rohit", "Neha", "Priya", "Sonal", "Rohit", "Priya", "Arjun", "Kavita", "Neha"],
    "Course": ["B.Tech", "B.Sc", "B.Tech", "MBA", "BBA", "B.Sc", "MBA", "B.Tech", "MBA", "B.Tech"],
    "Marks": [85, 78, 92, 88, 70, 78, 88, 95, 80, 92]
}

df = pd.DataFrame(data)
print(df)
```

```python
df.duplicated()
df[df.duplicated()]
```

```python
df.drop_duplicates()
```

```python
df.drop_duplicates(keep="first")   
df.drop_duplicates(keep="last")    
```

```python
df[df.duplicated(subset="Roll_No")]
```

```python
df.drop_duplicates(subset="Roll_No", keep="first")
```

```python
df["is_duplicate"] = df.duplicated()
print(df)
```

```python
df[df.duplicated(keep=False)]
```

```python
df.groupby("Name").size()
```

```python
df.drop_duplicates(keep=False)
```

```python
df.groupby(["Roll_No", "Name", "Course"], as_index=False)["Marks"].mean()
```

```python
dupes_by_roll = df[df.duplicated(subset=["Roll_No"], keep=False)]
print(dupes_by_roll)
```
