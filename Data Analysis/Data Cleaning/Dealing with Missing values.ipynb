 In the field of data-related research, it is very important to handle missing data either by deleting or imputation(handling the missing values with some estimation).

Different Methods of Dealing With Missing Data

1. Deleting the column with missing data

2. Deleting the row with missing data

3. Filling the Missing Values – Imputation

   (i) Numerical data - use mean

   (ii) categorical data - use mode - assign the NaN values their own category

4. Advanced Imputation

```python
import pandas as pd 
```

```python
df=pd.read_csv('AB_NYC_2019.xls')
```

```python
df.head()
```

## **Deleting the column with missing data**

```python
df1=df.copy()
```

```python
df1.isna().sum()
```

```python
df1.drop("last_review",axis=1, inplace= True)
```

```python
df1.isna().sum()
```

## **Deleting the row with the missing dataa**

```python
df2=df1.copy()
```

```python
df2.isna().sum()
```

```python
df2.dropna(inplace =True )
```

```python
df2.isna().sum()
```

```python
df2.shape
```

## **Filling the Missing Values- Imputation**
  (i)Numerical data- use mean

```python
df3=df.copy()
```

```python
df3.head()
```

```python
mean_value = df3["reviews_per_month"].mean()
df3["reviews_per_month"] = df3["reviews_per_month"].fillna(mean_value)
```

```python
df3.head()
```

```python
df3.isna().sum()
```

## **Filling Categorical Data**

```python
df3.nunique()
```

```python
df3["reviews_per_month"].fillna(df3["reviews_per_month"].value_counts().index[0],inplace=True)
```

```python
df3.isna().sum()
```

```python
df4=df.copy()
```

```python
df4.isna().sum()
```

```python
df3["reviews_per_month"].fillna("Not_reviews", inplace = True)
```

```python
df4.sample(10)
```

## **Advanced Imputation**

```python
df5=df.copy()
```

```python
df5.isna().sum()
```

```python
df5["reviews_per_month"] = df5["reviews_per_month"].interpolate()
```
